```python
import pandas as pd 
import numpy as np 
import joblib 
import os
```

```python
from draft.data.sets import load_sets
```

```python
project_dir = os.path.dirname(os.getcwd()) 
data_dir = os.path.join(project_dir, "data") 
sets_dir = os.path.join(data_dir, "processed") 

X_train, X_val, X_test, y_train, y_val, y_test = load_sets(
    load_dir=sets_dir, 
    return_dict=False
)
```

```python
import xgboost as xgb
```

```python
xgboost1 = xgb.XGBClassifier() 
```

```python
from draft.model.eval import fit_assess_classifier 
```

```python
X_train[0]
```

```python
xgboost1, xg1_scores = fit_assess_classifier(
    xgboost1,
    X_train,
    y_train,
    X_val,
    y_val,
    metrics={"accuracy_score": {}, 
             "f1_score": {"average": "weighted"}} 
)
```

```python
models_dir = os.path.join(project_dir, "models") 
joblib.dump(xgboost1, os.path.join(models_dir, "xgboost_default.joblib"))
```

```python
xgboost2, xgb2_scores = fit_assess_classifier(
    xgb.XGBClassifier(max_depth=5), 
    X_train, 
    y_train, 
    X_val, 
    y_val, 
    metrics={"accuracy_score": {}, "f1_score": {"average": "weighted"}}
    ) 
```

```python
default_params = {
    'X_train': X_train,
    'y_train': y_train, 
    'X_val': X_val, 
    'y_val': y_val, 
    'metrics': {
        'accuracy_score': {}, 
        'f1_score': {
            'average': "weighted"
        }
    }
}

xgboost3, xg3_scores = fit_assess_classifier(
    xgb.XGBClassifier(max_depth=3), 
    **default_params
)
```
