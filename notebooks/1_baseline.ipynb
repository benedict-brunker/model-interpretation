## Data Preparation

```python
df_cleaned = df.copy()
```

```python
from draft.features.dates import convert_to_date
```

```python
df_cleaned = convert_to_date(df_cleaned, ["tpep_pickup_datetime", "tpep_dropoff_datetime"])
```

```python
df_cleaned[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].head()
```

```python
duration_secs = (df_cleaned['tpep_dropoff_datetime'] - df_cleaned['tpep_pickup_datetime']).dt.total_seconds() 

duration_secs
```

```python
df_cleaned['trip_duration'] = duration_secs
df_cleaned['trip_duration'].describe()
```

Convert the trip duration columns into 5 different bins with [0, 300, 600, 1800, 300000]

```python
trip_duration_bins = pd.cut(
    x=df_cleaned['trip_duration'], 
    bins=[-1, 300, 600, 1800, 300000],
    labels=[0, 1, 2, 3]
)
trip_duration_bins.describe()
```

```python
df_cleaned['trip_duration'] = trip_duration_bins
```

```python
df_cleaned['tpep_pickup_month'] = df_cleaned['tpep_pickup_datetime'].dt.month 
df_cleaned['tpep_pickup_month'].describe()
```

```python
df_cleaned['tpep_pickup_datetime'].describe()
```

```python
df_cleaned['tpep_pickup_hourofday'] = df_cleaned['tpep_pickup_datetime'].dt.hour 
df_cleaned['tpep_pickup_hourofday'].describe()
```

```python
df_cleaned['tpep_pickup_dayofweek'] = df_cleaned['tpep_pickup_datetime'].dt.weekday
df_cleaned['tpep_pickup_dayofweek'].describe()
```

```python
df_cleaned.columns
```

```python
from sklearn.preprocessing import OneHotEncoder 

ohe = OneHotEncoder(drop='first', sparse_output=False)
ohe.set_output(transform='pandas') 

ohm = ohe.fit_transform(df_cleaned[['VendorID', 'RatecodeID', 'store_and_fwd_flag']]) 

ohm.head()
```

```python
df_cleaned.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'PULocationID', 'DOLocationID'], axis=1, inplace=True) 
df_cleaned.columns
```

```python
datasets_dir = os.path.join(data_dir, "interim") 

df_cleaned.to_csv(os.path.join(datasets_dir, "df_cleaned.csv"), index=False) 

os.listdir(datasets_dir)
```

## Data Splitting

```python
from draft.data.sets import split_sets_random 

target = df_cleaned.pop('trip_duration') 

X_train, X_val, X_test, y_train, y_val, y_test = split_sets_random(
    df_cleaned, 
    target, 
    test_ratio=0.2, 
    seed=5, 
    as_type='numpy',
    ret_dict=False 
) 
```
